```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

path = "../local/Tvarminne/CH4.csv"

df = pd.read_csv(path, engine="pyarrow")
df["Time"] = pd.to_datetime(df["Time"], format="%d-%b-%Y %H:%M:%S", errors='coerce')
df["FCH4_nmolm2s"] = pd.to_numeric(df["FCH4Pi"])
df = df[["Time", "FCH4_nmolm2s"]]
df.set_index("Time", inplace=True)

print(df.index.dtype)
print(df.head())

df.to_csv("../local/Tvarminne/preprocessed_all/ch4.csv", index=True, date_format="%Y-%m-%d %H:%M:%S")
```
