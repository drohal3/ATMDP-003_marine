```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

path = "../local/Tvarminne/MSA.xlsx"

df = pd.read_excel(path)

df["Time"] = pd.to_datetime(df["Time"], format="%d-%b-%Y %H:%M:%S", errors='coerce')
df.set_index("Time", inplace=True)

for col in df.columns:
    print(col)
    df[col] = df[col].mask(df[col] < 0, np.nan) 

df = df.resample("h").median()

df.to_csv("../local/Tvarminne/preprocessed_all/msa_preprocessed.csv", index=True, date_format="%Y-%m-%d %H:%M:%S")
```
