```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

path = "../local/Tvarminne/Vocus_merged.csv"

df = pd.read_csv(path, engine="pyarrow")
df["Time"] = pd.to_datetime(df["Time"], format="%d-%b-%Y %H:%M:%S", errors='coerce')
df.set_index("Time", inplace=True)

for col in df.columns:
    df[col] = df[col].mask(df[col] <= 0, np.nan) 

df = df.resample("h").median()

print(df.dtypes)
print(df.index.dtype)
print(df.head())
print(df.shape)

df.to_csv("../local/Tvarminne/preprocessed_all/vocs_many.csv", index=True, date_format="%Y-%m-%d %H:%M:%S")


# DO NOT FORGET TO AGGREGATE!!!
```
